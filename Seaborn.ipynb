```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Qn1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
sns.relplot(x='total_bill', y='tip', data = df, hue='smoker')
```

```python
# Qn2: Generate a dataset of random numbers. Visualize the distribution of a numerical variable.
```

```python
data = np.random.randn(100)
sns.histplot(data, kde=True)
plt.title('Distribution of random numbers')
plt.xlabel('values')
plt.ylabel('Frequency')
plt.show()
```
